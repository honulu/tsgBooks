```python
from bs4 import BeautifulSoup
```

```python
r = open("C:\\Users\\honulu\\Desktop\\code\\python\\tsgBooks\\data\\test.html","r",encoding="utf-8")
k = r.read()
r.close()
```

```python
soup = BeautifulSoup(k)
# print(soup.prettify())
print(soup.text)
```

```python
import re
k = soup.findAll("div", "list_books")
for i in k:
    # books's name
    print( re.sub(r'\d*\. ',"",i.h3.get_text()))
    # book's website
    print(re.sub(r"\D", "", i.p.span["id"]))
    # book's publisher
    print(re.sub(r"\s", "", i.p.get_text()))
```
